```python
import pandas as pd
```

```python
dados_vendas = pd.read_json('/content/drive/MyDrive/Colab Notebooks/Alura/Formação Data Science Python/7_transformacao_e_manipulacao_de_dados/desafios/desafio_1/dados_vendas_clientes.json')
dados_vendas.head()
```

```python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
dados_vendas.head()
```

```python
colunas = list(dados_vendas.columns)
colunas
```

```python
dados_vendas = dados_vendas.explode(colunas[1:])
dados_vendas
```

```python
dados_vendas.reset_index(inplace=True, drop=True)
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas['Valor da compra']
```

```python
import numpy as np
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip())
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].astype(np.float64)
```

```python
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas.head()
```

```python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower()
```

```python
dados_vendas.head()
```

```python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.replace('[^a-z ]', '', regex=True)
```

```python
dados_vendas
```

```python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.strip()
```

```python
dados_vendas.head()
```

```python
dados_vendas['Data de venda'] = pd.to_datetime(dados_vendas['Data de venda'], format='%d/%m/%Y')
```

```python
dados_vendas
```

```python
total_compras = dados_vendas.groupby(['Cliente'])['Valor da compra'].sum()
```
