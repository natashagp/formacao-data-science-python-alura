```python
import pandas as pd
```

```python
dados_condominios = pd.read_json('/content/drive/MyDrive/Colab Notebooks/Alura/Formação Data Science Python/7_transformacao_e_manipulacao_de_dados/desafios/desafio_2/dados_locacao_imoveis.json')
dados_condominios.head()
```

```python
dados_condominios = pd.json_normalize(dados_condominios['dados_locacao'])
dados_condominios.head()
```

```python
colunas = list(dados_condominios.columns)
colunas
```

```python
dados_condominios = dados_condominios.explode(colunas[1:])
```

```python
dados_condominios
```

```python
dados_condominios.reset_index(inplace=True, drop=True)
```

```python
dados_condominios
```

```python
dados_condominios.info()
```

```python
dados_condominios['valor_aluguel']
```

```python
import numpy as np
```

```python
dados_condominios['valor_aluguel'] = dados_condominios['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace('reais', '').replace(',', '.').strip())
```

```python
dados_condominios['valor_aluguel'] = dados_condominios['valor_aluguel'].astype(np.float64)
```

```python
dados_condominios
```

```python
dados_condominios.info()
```

```python
dados_condominios.head()
```

```python
dados_condominios['apartamento'] = dados_condominios['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
```

```python
dados_condominios.head()
```

```python
dados_condominios['datas_combinadas_pagamento'] = pd.to_datetime(dados_condominios['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_condominios['datas_de_pagamento'] = pd.to_datetime(dados_condominios['datas_de_pagamento'], format='%d/%m/%Y')
```

```python
dados_condominios.info()
```

```python
dados_condominios['atraso'] = (dados_condominios['datas_de_pagamento'] - dados_condominios['datas_combinadas_pagamento']).dt.days
```

```python
dados_condominios
```

```python
media_atraso = dados_condominios.groupby(['apartamento'])['atraso'].mean()
media_atraso
```
