Você recebeu a tarefa de normalizar o seguinte arquivo JSON chamado “informacoes.json” referente a algumas informações de identificação de um cliente.

```python
import pandas as pd
```

```python
import json
```

```python
with open("/content/drive/MyDrive/Colab Notebooks/Alura/Formação Data Science Python/8_pandas_limpeza_e_tratamento_de_dados/desafios/desafio_1/informacoes.json") as f:
    json_bruto = json.load(f)
```

```python
informacoes = pd.json_normalize(json_bruto, record_path='enderecos', meta=['nome', 'idade'])
```
