Sua tarefa é realizar alguns tratamentos e deixar esse arquivo JSON como um Pandas DataFrame, sem dados vazios e com a tipagem das colunas de forma correta. Como isso pode ser feito?

```python
import json
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/Alura/Formação Data Science Python/8_pandas_limpeza_e_tratamento_de_dados/desafios/desafio_2/dados.json', 'r') as f:
    dados = json.load(f)
```

```python
import pandas as pd
```

```python
# Normalizar o Json
df = pd.json_normalize(dados, record_path=['pessoas'])
```

```python
df
```

```python
df.info()
```

```python
# Cast da coluna idade para int
df['idade'] = df['idade'].astype(int)
```

```python
df.info()
```

```python
# Filtro de valores que não são vazios
df_filtrado = df.query('`endereco.rua` != ""')
df_filtrado
```

```python
df_filtrado = df[df['endereco.rua'] != ""]
```
